```python
def printTree(root):
    
    # this is not a base case its just an edge case to handle incase the overall tree has no node 
    # in case of a not-null tree we do not even need these line 
    #loop through the children nodes itself will take care of the base case and recursion works fine 
    
    if root == None:
        print("I am Hit !!!")
        return 
    
    print(root.val)
    
    for ele in root.noChildren:
        printTree(ele)
```

```python
def printTree(root):
    
    # this is not a base case its just an edge case to handle incase the overall tree has no node 
    # in case of a not-null tree we do not even need these line 
    #loop through the children nodes itself will take care of the base case and recursion works fine 
    
    if root == None:
        print("I am Hit !!!")
        return
        
    print(root.val, end = ":")
    
    for kids in root.noChildren:
        print(kids.val, end = ",")
    print()
    for ele in root.noChildren:
        printTree(ele)
```

### Taking input from user to build the tree 

```python
def userInputRecursive():
    
    rootVal =  int(input("Nodes value : "))
    
    if rootVal == -1:
        return None
    
    root = TreeNode(rootVal)
    noChild = int(input(f"Please enter the No of children for {root.val} : "))
    
    for i in range(noChild):
        print(f"Enter val for the {i}th child of {root.val} : ")
        Node = userInputRecursive()
        root.noChildren.append(Node)
        
    return root
```

```python
root = userInputRecursive()
printTree(root)
```

### level wise entry of generic tree nodes :

                We will have to use a queue to implement "level wise input for Tree !!!"

```python
import queue

def treeInputlevelWise():
    
    val = int(input(f"Node's Value : "))
    
    if val == -1:
        return 
    
    q = queue.Queue()
    
    root = TreeNode(val)
    q.put(root)
    
    while q.empty()!= True:
        poppedNode = q.get()
        noChildren = int(input(f"How many children you have for {poppedNode.val} : "))
        
        for i in range(noChildren):
            val = int(input(f"\tChild Number  {i} : "))
            newNode = TreeNode(val)
            poppedNode.noChildren.append(newNode)
            q.put(newNode)
    return root
```

```python
root = treeInputlevelWise()
print('\n\n')
printTree(root)
```

### Find the Number of Nodes in a Generic Tree :

```python
def noNodes(root):
    
    if root == None:
        return 0
    
    nodeCount = 1
    
    for childNodes in (root.noChildren):
        localCount = noNodes(childNodes)
        nodeCount  += localCount
    
    return nodeCount
```

```python
print(noNodes(root))
```
