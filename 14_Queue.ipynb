### Queue using Lists/Arrays: 

```python
# Whenever I deque then I +1 the front index so that Dequeue is acheived in O(1) time
# For size maintain a count

class QueueUsingArray:
    def __init__(self):
        self.__arr = []
        self.__count = 0
        self.__front = 0
        
    def enqueue(self, data):
        self.__arr.append(data)
        self.__count += 1
        
    def dequeue(self):
        if self.__count == 0:
            return -1
        
        element = self.__arr[self.__front]
        self.__front += 1
        self.__count -= 1
        return element
        
    def front(self):
        if self.__count == 0:
            return -1
        return self.__arr[self.__front]
        
    def size(self):
        return self.__count
    
    def isEmpty(self):
        return self.size() == 0
```

```python
q = QueueUsingArray()

q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)

while q.isEmpty() is False:
    print(q.front())
    q.dequeue()
    
print(q.dequeue())
```

### Queue using Linked Lists 

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        

class QueueUsingLL:
    
    def __init__(self):
        self.__head = None
        self.__tail = None
        self.__count = 0
  
        
    def enqueue(self, element):
        newNode = Node(element)
        if self.__head is None:
            self.__head = newNode            
        else:
            self.__tail.next = newNode
            
        self.__tail = newNode
        self.__count += 1
            
        
    def dequeue(self):        
        if self.__head is None:
            print("Hey! Queue is Empty")
            return 
        
        ele = self.__head.data
        self.__head = self.__head.next
        self.__count -= 1
        
        return ele
        
    def front(self):
        if self.__count == 0:
            print("Queue is Empty!")
            return 
        
        return self.__head.data
        
    def size(self):
        return self.__count
    
    def isEmpty(self):
        return self.size() == 0
    
```

```python
q = QueueUsingLL()

q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)

print(q.size())
print(q.isEmpty())
print(q.front())

while q.isEmpty() is not True:
    print(q.dequeue())
print(q.front())
```

```python
class Stack(object):
    
    def __init__(self):
        self.__data = []
        
    def printstack(self):
        for ele in self.__data:
            print(ele,end = ' ')
        
        return
        
    def push(self,val):
        self.__data.append(val)
        
    def pop(self):
        if self.isEmpty():
            print("stack is empty !!!")
            return
        return self.__data.pop()
    
    
    def top(self):
        if self.isEmpty():
            print("stack is empty !!!")
            return
        return self.__data[len(self.__data)-1]
        
    def size(self):
        return len(self.__data)
        
    def isEmpty(self):
        return self.size() == 0
```

```python
class Queue:
    
    def __init__(self):
        self.s1 = Stack()
        self.s2 = Stack()
    
    
    def additem(self,val):
        self.s1.push(val)
        self.s2.push(val*2)
    
    def printQ(self):
        self.s1.printstack()
        print('\n')
        self.s2.printstack()
```

```python
myQ = Queue()
myQ.additem(5)
myQ.additem(52)
myQ.printQ()
```

```python
class Queue:
    
    def __init__(self):
        self.s1 = Stack()
        self.s2 = Stack()
    
    
    def enqueue(self,val):
        self.s1.push(val)
    
    
    def dequeue(self):
        
        if self.s1.size() == 0 :
            return 
        
        while self.s1.size() != 1 :
            temp = self.s1.pop()
            self.s2.push(temp)
        
        x = self.s1.pop()
    
        while self.s2.size() != 0 :
            temp = self.s2.pop()
            self.s1.push(temp)
            
        return x
    
    def printQ(self):
        self.s1.printstack()
        
    def size(self):
        return self.s1.size()
    
    def isEmpty(self):
        return self.s1.isEmpty()
```

```python
myQ = Queue()
myQ.enqueue(5)
myQ.enqueue(2)
myQ.enqueue(7)
myQ.enqueue(13)
myQ.enqueue(8)
myQ.enqueue(1)

myQ.printQ()

item = myQ.dequeue()
print()
print("Dequeued Item : ",item)
item = myQ.dequeue()
print("Dequeued Item : ",item)

myQ.printQ()
```
