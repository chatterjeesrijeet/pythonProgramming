### Binary Search using Recursion :
            Condition :  List must be Sorted 

```python
def binarySearch(myList,ele):
    
    if len(myList) == 0 :
        return -1
    
    mid = len(myList)//2
    
    if myList[mid] == ele:
        return mid
    else:
        #make a smaller list with half the elements
        if ele > myList[mid]:
            newList = myList[mid+1:]
        else:
            newList = myList[:mid]
        
        #call the binary function on smaller problem 
        smallerListIndex = binarySearch(newList,ele)
        
        #depending on the smaller solution create teh final solution 
        if smallerListIndex == -1:
            return -1
        elif ele > myList[mid]:
            return mid + (smallerListIndex+1)
        else:
            return smallerListIndex
        
```

### Merge Sort

```python
def mergingSortedLists(listA, listB):
    
    currIndexA, currIndexB = 0, 0
    sortedList = []
    
    while (currIndexA < len(listA)) and (currIndexB < len(listB)):
        
        if listA[currIndexA] <= listB[currIndexB]:
            sortedList.append(listA[currIndexA])
            currIndexA += 1
        else:
            sortedList.append(listB[currIndexB])
            currIndexB += 1
            
    #if List A is completed first 
    if currIndexA == len(listA):
        while currIndexB != len(listB):
            sortedList.append(listB[currIndexB])
            currIndexB += 1
    
    #if List B is completed first         
    else:
        while currIndexA != len(listA):
            sortedList.append(listA[currIndexA])
            currIndexA += 1
    
    return sortedList
```

### Quick Sort

```python
def partition(arr,start,end):
    
    count = 0
    pivot_element = arr[end]
    for index in range(start,end):
        if arr[index] <= pivot_element:
            count += 1
    
    pivot_index = start+count
    
#     if pivot_index == end:
#         return
    
    arr[pivot_index],arr[end] = arr[end],arr[pivot_index]
    
    for i in range(start,pivot_index):
        if arr[i] > pivot_element:
            for j in range(end,pivot_index,-1):
                if arr[j]<= pivot_element:
                    arr[i],arr[j] = arr[j],arr[i]
                    break
            
    return pivot_index
```

```python
def quickSort(arr, start, end):
    if start >= end:
        return -1
    index = partition(arr,start,end)
    quickSort(arr,start,index-1)
    quickSort(arr,index+1,end)
```

### Tower of Hanoi

```python
'''
a = source_tower 
c = destination_tower
b = helper_tower 

(no_of_disks,source_tower --> destination_tower --->  via helper_tower)

move a to c using b

'''

def towerOfHanoi(n,a,c,b):
    
    if n == 1 :
        print(f"Move Disk_No : {n}, {a} ---> {c}")
        return
    
    towerOfHanoi(n-1,a,b,c)
    print(f"Move Disk_No : {n}, {a} ---> {c}")
    towerOfHanoi(n-1,b,c,a)
```

### Finding all sub sequences of a string :

```python
def subsequences(string):
    #Implement Your Code Here
    if len(string) == 0:
        return ['']
    
    #lets get the solution from rest of the part recursively 
    smallerSolution = subsequences(string[1:])
    
    #construct my solution from the recursively received solution 
    myList = []
    for ele in smallerSolution:
        myList.append(string[0]+ele)
    smallerSolution.extend(myList)
    
    return smallerSolution


string = input()
ans = subsequences(string)

print("Length of Sub Sequences : ",len(ans))
for ele in ans:
    print(ele)
```

### Return Keypad Combinations :

```python
keypad_dict = {'2':'abc','3':'def','4':'ghi','5':'jkl','6':'mno','7':'pqrs','8':'tuv','9':'wxyz'}
```

```python
def returnKeypad(n):
    
    if n == 0 :
        return ['']
    
    smallNumber = n//10
    reminder = n%10
    
    smallerSolutionList = returnKeypad(smallNumber)
    
    string = keypad_dict[str(reminder)]
    
    solutionList = []
    
    for x in string:
        for ele in smallerSolutionList:
            solutionList.append(ele+x)
    return solutionList
```

### Return Permutations of a String

```python
def permutations(s):
    #Implement Your Code Here
    if len(s)==1:
        return [s]
    
    smallSol = permutations(s[1:])
    charathand = s[0]
    
    finalSolution = []
    
    
    for ele in smallSol:
        currstring = ele
        for i in range(len(currstring)+1):
            pre = currstring[:i]
            post = currstring[i:]
            tstring = pre+charathand+post
            finalSolution.append(tstring)
        
    return finalSolution    


string = input()
ans = permutations(string)
for s in ans:
    print(s)
```
