## Bubble Sort :

```python
def bubSort(L):
    
    i = 1
    while i <= len(L)-1:
        for j in range(0,len(L)-i):
            if L[j]>L[j+1]:
                L[j],L[j+1] = L[j+1],L[j]
        i += 1
        print(f'\nAfter Round {i}, L : {L}')
```

```python
n = int(input("Please Enter the number of Elements in the array : "))
L = [int(i) for i in input('Array element Vals : ').strip().split()]

index = bubSort(L)
```

```python
print(f"\nThe sorted list : {L} ")
```

    Lets modify it with a flag to ensure that incase there are no swap in a complete swipe then we can directly break.

```python
def bsort(l):
    
    for i in range(len(l)-1):
        swap = False
        for j in range(0,len(l)-1-i):
            if l[j]>l[j+1]:
                l[j],l[j+1] = l[j+1],l[j]
                swap = True
        if swap == False:
            break
    return l
```

## Selection Sort

```python
def selSort(l):
    
    sortedList = []
    
    #if n elements we need (n-1) rounds
    for i in range(1,len(l)):
        
        starting_index = i-1
        
        for j in range(starting_index+1,len(l)):
            if l[j] < l[starting_index]:
#                 print(f'true for {j}')
                l[starting_index],l[j] = l[j],l[starting_index]  
```

```python
n = int(input("Please Enter the number of Elements in the array : "))
L = [int(i) for i in input('Array element Vals : ').strip().split()]

index = selSort(L)
```

```python
print(f"\nThe sorted list : {L} ")
```

```python
def selSortCorrect(l):
    
    #if n elements we need (n-1) rounds
    for i in range(len(l)-1):
        
        min_index = i        
        for j in range(min_index+1,len(l)):
            if l[j] < l[min_index]:
                min_index = j 
        l[i],l[min_index] = l[min_index],l[i]
```

```python
n = int(input("Please Enter the number of Elements in the array : "))
L = [int(i) for i in input('Array element Vals : ').strip().split()]

index = selSortCorrect(L)
```

```python
print(f"\nThe sorted list : {L} ")
```

### Insersion Sort

```python
def insertionSort(arr):
 
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
 
        key = arr[i]
 
        # Move elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i-1
        while j >= 0 and key < arr[j] :
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key
 
 
# Driver code to test above
arr = [12, 11, 13, 5, 6]
insertionSort(arr)
for i in range(len(arr)):
    print ("% d" % arr[i])
```

## Merge Sort

```python
def mergingSortedLists(listA, listB):
    
    currIndexA, currIndexB = 0, 0
    sortedList = []
    
    while (currIndexA < len(listA)) and (currIndexB < len(listB)):
        
        if listA[currIndexA] <= listB[currIndexB]:
            sortedList.append(listA[currIndexA])
            currIndexA += 1
        else:
            sortedList.append(listB[currIndexB])
            currIndexB += 1
            
    #if List A is completed first 
    if currIndexA == len(listA):
        while currIndexB != len(listB):
            sortedList.append(listB[currIndexB])
            currIndexB += 1
    
    #if List B is completed first         
    else:
        while currIndexA != len(listA):
            sortedList.append(listA[currIndexA])
            currIndexA += 1
    
    return sortedList
```

```python
print( mergingSortedLists(listA = [14], listB = [12]))
```

```python
def myMergeSort(myList):
    
    if len(myList) <= 1 :
        return myList
    
    mid = (len(myList))//2
    
    listA = myList[0:mid]
    listB = myList[mid:]
    
    sortedListA = myMergeSort(listA)
    sortedListB = myMergeSort(listB)
    
    myList = mergingSortedLists(sortedListA, sortedListB)
    
    
    return myList
```

```python
print(myMergeSort([1,3,2,12,23,0,24,23,11,18]))
```

## Quick Sort 

```python
def partition(arr,start,end):
    
    count = 0
    pivot_element = arr[end]
    for index in range(start,end):
        if arr[index] <= pivot_element:
            count += 1
    
    pivot_index = start+count
    
    arr[pivot_index],arr[end] = arr[end],arr[pivot_index]
    
    for i in range(start,pivot_index):
        if arr[i] > pivot_element:
            for j in range(end,pivot_index,-1):
                if arr[j]<= pivot_element:
                    arr[i],arr[j] = arr[j],arr[i]
                    break
            
    return pivot_index
```

```python
# arr = [88,21,73,18,22,119,13,22,90,-13,11,27]
arr = [11,23,78]
x = partition(arr,0,len(arr)-1)
print(x)
print(arr)
```

```python
def quickSort(arr, start, end):
    if start >= end:
        return -1
    index = partition(arr,start,end)
    quickSort(arr,start,index-1)
    quickSort(arr,index+1,end)
```

```python
n = int(input())
arr = list(int(i) for i in input().strip().split(' '))
quickSort(arr, 0, n-1)
print(arr)
```

### 2 D Lists 

```python
x = input().strip().split()
n,m = int(x[0]),int(x[1])

L = [ [ int(x) for x in input().strip().split()] for i in range(n)]
```

```python
def colSumIndex(L,n,m):
    
    col_sum = -99
    col_index = -1
    
    for j in range(m):
        sum = 0
        for i in range(len(L)):
            sum += L[i][j]
        if sum > col_sum:
            col_sum = sum 
            col_index = j
    return col_index
```

```python
colSumIndex(L,n,m)
```
