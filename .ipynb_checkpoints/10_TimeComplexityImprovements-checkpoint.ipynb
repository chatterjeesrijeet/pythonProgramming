#### Approach 1 :

```python
def fapproach1(arr):
    
    for i in range(len(arr)):
        found = True
        for j in range(len(arr)):             ##        O(n^2)
            if i==j:
                continue
            if arr[i] == arr[j]:
                found = False
                break
        if found == True:
            
            return arr[i]
    return None
```

```python
print(fapproach1([1,1,2,2,3,3,4,4,5]))
```

#### Approach 2:  

```python
# usinga sorted array and compare with immediate next element 
def fapproach2(arr):
    arr.sort()
    print(arr)
    n = (len(arr)//2)*2                          ##        O(nlogn)
    for i in range(0,n,2):
        if arr[i] != arr[i+1]:
            return arr[i]
    return arr[len(arr)-1]
```

```python
print(fapproach2([1,2,3,2,1,3,4]))
```

```python
print(fapproach2([1,2,2,1,3,4]))
```

#### Approach 3 :

```python
def fapproach3(arr):
    temp = 0
    for i in range(len(arr)):                    ##        O(n)
        temp = temp^arr[i]
    return temp
```

```python
print(fapproach3([1,2,3,2,1,3,4]))
```

#### Approach 1 :

```python
def f1(arr):
    for i in range(len(arr)):
        index = i 
        leftarraysum = sum(arr[0:index])         ## O(n^2)
        rightsum = sum(arr[index+1:])
        if leftarraysum == rightsum:
            return i
```

```python
print(f1([1,2,3,9,4,7,6,2]))
```

#### Approach 2 :

```python
def f2(arr):
    leftsum = 0
    rightsum = sum(arr)                       ## O(n)
    for i in range(len(arr)):
        rightsum = rightsum-arr[i]  ## for the right sum we subtract from new index i 
        if leftsum == rightsum:
            return i
        leftsum = leftsum + arr[i]
        
    return None
```

```python
print(f2([1,2,3,9,4,7,6,2]))
```

#### Approach 1 :

    Time complexity : O(n * d) 
    Auxiliary Space : O(1) 

```python
def f1(arr,d):
    while d>0:
        temp = arr[0]
        for i in range(0,len(arr)-1):
            arr[i] = arr[i+1]
        
        arr[len(arr)-1]= temp
        
        d = d-1
        
```

```python
arr = [1, 2, 3, 4, 5, 6, 7]
d = 2
f1(arr,d)
print(arr)
```

#### Approach 2 :
          Time complexity: O(n) 
          Auxiliary Space: O(d)

```python
def f2(arr,d):
    
    temp = arr[0:d]
    j = 0 
    for i in range(0,len(arr)):
        if i < len(arr)-d:
            arr[i] = arr[i+d]
        else:
            arr[i] = temp[j]
            j = j+1
```

```python
arr = [1, 2, 3, 4, 5, 6, 7]
d = 2
f2(arr,d)
print(arr)
```

#### Approach 3 :
         Time complexity: O(n) 
         Auxiliary Space: O(1)

```python
def mySwap(arr,i,j):
    while i<j:
        arr[i],arr[j] = arr[j],arr[i]
        i += 1
        j -= 1
```

```python
arr = [1, 2, 3, 4, 5, 6,7]
mySwap(arr,0,len(arr)-1)
print(arr)
arr = [1, 2, 3, 4, 5, 6]
mySwap(arr,0,len(arr)-1)
print(arr)
```

```python
def f3(arr,d):
    mySwap(arr,0,len(arr)-1)
    mySwap(arr,0,len(arr)-1-d)
    mySwap(arr,len(arr)-d,len(arr)-1)
```

```python
arr = [1, 2, 3, 4, 5, 6, 7]
d = 2
f3(arr,d)
print(arr)
```

#### Approach 4 :
            Time complexity : O(n) 
            Auxiliary Space : O(1)

```python
def f4(arr,d):
    return arr[d:]+arr[0:d]
```

```python
arr = [1, 2, 3, 4, 5, 6, 7]
d = 2
arr = f4(arr,d)
print(arr)
```

```python
arr = [1, 2, 3, 4, 5, 6, 7]
d = 7
arr = f4(arr,d)
print(arr)
```
