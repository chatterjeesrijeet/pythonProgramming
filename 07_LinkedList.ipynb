### Taking Input from User and stop at -1 

       1. Option 1 : Take everything and then create --------> Most Common and used in Coding rounds
       2. Option 2 : create in the run time till user enters -1  

```python
class Node:
    
    def __init__(self, data):
        self.data = data
        self.next = None
```

      O(n**2) time complexity code

```python
def user_input():
    
    user_data = [int(x) for x in input().strip().split()]
    head = None
    
    for data in user_data:

        if data == -1 :
            break 

        newNode = Node(data)

        if head==None:
            head = newNode
        else:
            curr = head
            while curr.next != None:
                curr = curr.next
            curr.next = newNode
    return head
```

## Inserting at an index

```python
class Node(object):
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
def user_input():
    l = [int(x) for x in input().strip().split()]
    
    head = None
    tail = None
    
    for ele in l:
        if ele == -1 :
            break
        else:
            newNode = Node(ele)
            if head is None:
                head = newNode
                tail = newNode
            else:
                tail.next = newNode
                tail = tail.next
    return head
```

```python
def printll(head):
    if head is None:
        print("Linked List is empty !!!!")
    else:
        while head != None:
            print(head.data,end = '--->')
            head = head.next
        print("None")
```

```python
def lenll(head):
    
    count = 0 
    
    while head != None:
        count += 1
        head = head.next
    return count
```

```python
def insertll(head,index,val):
    
    if index < 0 or index > lenll(head):
        print("Index is out of bounds !!!")
        return head
    
    curr = head
    prev = None
    
    i = 0 
    
    while i != index:
        i = i + 1
        prev = curr
        curr = curr.next
    
    newNode = Node(val)
    newNode.next = curr
    
    if i == 0 :
        head = newNode
    else:
        prev.next = newNode
    
    return head
```

```python
head = user_input()
printll(head)

val = int((input()))
index = int(input())

head = insertll(head,index,val)
printll(head)
```

```python
def insertllRec(head,val,index):
    
    if head == None or index < 0:
        print('\nIndex out of bounds, please check !!! \n')
        return head
    
    if index == 0:
        newNode = Node(val)
        newNode.next = head
        head = newNode
        return head
    
    
    newHead = insertllRec(head.next,val,index-1)
    
    head.next = newHead
    
    return head
```

```python
head = user_input()
print("\nThe Linked List that you entered :\n\n ")
printll(head)

val = int((input("\nEnter the value you want to enter : ")))
index = int(input("\nIndex position where you want to enter : "))

head = insertllRec(head,val,index)
print("\nThe modified Linked List after your insersion :\n\n ")
printll(head)
```

# Deleting the ith node  of a Linked List :

      ---------------------------------------- Iterative Code --------------------------------------------- 

```python
def delll(head,index):
    
    
    if index < 0 or index >= lenll(head):
        print("\n\tIndex out of bounds !!!")
        return head
    
    prev = None
    curr = head
    
    i = 0
    
    while i != index:
        i += 1
        prev = curr
        curr = curr.next
    
    if index==0:         #to remove 0th index we need to update the head pointer itself
        head = curr.next
    else:
        prev.next = curr.next
    
    curr.next = None
    
    return head
```

```python
head = user_input()
print("\nThe Linked List that you entered :\n")
printll(head)

#val = int((input("\nEnter the value you want to enter : ")))
index = int(input("\nIndex position which you want to delete : "))

head = delll(head,index)
print("\nThe modified Linked List after your deletion :\n\t")
printll(head)
```

```python
def delll_rec(head,index):
    
    if index < 0 or head == None:
        print("\n\tInvalid Index\n")
        return head
    
    if index == 0 :
        head = head.next
        return head
    
    smallhead = delll_rec(head.next,index-1)
    
    head.next = smallhead
    
    return head
```

```python
head = user_input()
print("\nThe Linked List that you entered :\n")
printll(head)

#val = int((input("\nEnter the value you want to enter : ")))
index = int(input("\nIndex position which you want to delete : "))

head = delll_rec(head,index)
print("\nThe modified Linked List after your deletion :\n\t")
printll(head)
```

###  Reverse The Linked List 

```python
def revLL(head):
    
    if lenll(head) <= 1 :
        return head
    
    reversedhead = revLL(head.next)
    
    tail = reversedhead
    while tail.next != None:
        tail = tail.next
        
    tail.next = head
    head.next = None
    
    return reversedhead
    
```

```python
head = user_input()
print("\nThe Linked List that you entered :\n")
printll(head)

head = revLL(head)

print("\nThe modified Linked List after reversal:\n\t")
printll(head)
```

```python
def revLL_optim(head,tail):
    
    if head == tail:
        return head,tail
    
    reversedhead,reversedtail = revLL_optim(head.next,tail)
    
    reversedtail.next = head
    head.next = None
    
    reversedtail = reversedtail.next
    
    return reversedhead,reversedtail
    
```

```python
def mergingsortedll(head1,head2):
    
    if head1==None or head2 == None:
        return head1 if head2 == None else head2
    
    v1,v2 = head1,head2
    minp = None
    
    # this is to place the minpointer at teh start of the algorithm and save the sorted head 
    if v1.data < v2.data :
        minp = v1
        sortedhead = v1
        v1 = v1.next             # this might make v1 as none 
    else:
        minp = v2
        sortedhead = v2
        v2 = v2.next             # this might make v2 as none 
    
    # lets now loop through the linked lists and 
    # find the min val of v1 and v2 pointers move the minp's next to that pointer and proceed further  
    
    while v1 != None and v2 != None:
        
        if v2.data < v1.data :
            minp.next = v2
            minp = v2
            v2 = v2.next
            
        else:
            minp.next = v1
            minp = v1
            v1 = v1.next
    
    # logic once one of them ends 
    if v2 == None:
        minp.next = v1
    else:
        minp.next = v2
        
    return sortedhead
```
