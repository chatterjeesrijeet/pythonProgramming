## Implementation of Stack using LL and Lists(Array)

         (A). Using Lists/Arrays

```python
class Stack(object):
    
    def __init__(self):
        self.__data = []
        
    def printstack(self):
        for ele in self.__data:
            print(ele,end = ' ')
        
        return
        
    def push(self,val):
        self.__data.append(val)
        
    def pop(self):
        if self.isEmpty():
            print("stack is empty !!!")
            return
        return self.__data.pop()
    
    
    def top(self):
        if self.isEmpty():
            print("stack is empty !!!")
            return
        return self.__data[len(self.__data)-1]
        
    def size(self):
        return len(self.__data)
        
    def isEmpty(self):
        return self.size() == 0
```

```python
class Node(object):
    
    def __init__(self,data):
        self.data = data
        self.next = None

        
class Stack(object):

    def __init__(self):
        self.__head = None
        self.__tail = None
        
    def push(self,val):
        
        newNode = Node(val)
        
        if self.__head == None:
            self.__head = newNode
            self.__tail = newNode
        else:
            self.__tail.next = newNode
            self.__tail = self.__tail.next
    
    def pop(self):
        
        if self.__head == None:
            print("\tCann't POP as the stack is empty !!!")
            return
        
        if self.__head == self.__tail:    #stack has only one element 
            self.__head = None
            self.__tail = None
        
        else:
            curr = self.__head
            
            while curr.next != self.__tail:
                curr = curr.next
            
            curr.next = None
            self.__tail = curr
            
    def stacktop(self):
        if self.__head == None:
            print("\tThe stack is empty so no top element!!!")
        else:
            print("\tThe Top element in the stack is : ",self.__tail.data)

    def size(self):
        
        if self.__head == None:
            print("\tThe size of the stack is : 0 ")
            return
        
        count = 1
        curr = self.__head
        
        while curr != self.__tail:
            curr = curr.next
            count += 1
        print("\tThe size of the stack is : ",count)
        
        return
    
    def isEmpty(self):
        return self.__head == None
        
        
        
    def printstack(self):
        
        if self.__head == None:
            print("\tCan not Print as the stack is empty !!!")
            return
            
        curr = self.__head
        
        while curr.next != None:
            print(curr.data)
            curr = curr.next
        print(curr.data)
```

### Balanced Parenthesis :

```python
def isBalanced(user_int):
    s = Stack()
    for char in user_int:
#         print(char)

        if char == '{' or char == '(' or char == '[':    #char in '{[('
            print("pushed : ",char)
            s.push(char)

        elif char == '}' or char == ')' or char == ']':   #char in ']})'
            if char == '}' and s.top()== '{':
                print("popped : ",char)
                s.pop()
            elif char == ']' and s.top()== '[':
                print("popped : ",char)
                s.pop()
            elif char == ')' and s.top()== '(':
                print("popped : ",char)
                s.pop()
            else:
                print("pushed : ",char)
                s.push(char)
                break
        else:
            continue

    if s.size()== 0:
        return True
    else:
        return False
```

```python
user_int = input("Enter the String : ")
status = isBalanced(user_int)
if status:
    print("\nThe expression has balanced parentheses !!!")
else:
    print("\nThe expression has im-balanced parentheses !!!")
```

### Reverse a stack using recursion & no additional space :

             not allowed to use loop constructs like while, for..etc, and you can only use the following ADT functions  
             on Stack S: 
                    1. isEmpty(S) /s.size()
                    2. push(S)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    
class Stack:
    def __init__(self):
        self.__head = None
        self.__cnt = 0
        
    def push(self, element):
        newNode = Node(element)
        newNode.next = self.__head
        self.__head = newNode
        self.__cnt += 1
        
    def pop(self):
        if self.isEmpty() is True:
            print("Hey! Stack is Empty!!")
            return
        
        data = self.__head.data
        self.__head = self.__head.next
        self.__cnt -= 1
        
        return data
        
    def top(self):
        if self.isEmpty() is True:
            #print("Hey! Stack is Empty!!")
            return
        
        return self.__head.data
    
    def size(self):
        return self.__cnt
        
    def isEmpty(self):
        return self.size() == 0         # return self.__head is None      // You can sue this as well
```

```python
def revStackRec(s):
    
    if s.size() == 0 or s.size()==1:
        return
    
    top_item = s.pop()          # save the top iten 
    revStackRec(s)              # send the rest of the stack  
    pushAtBottom(s,top_item)    # add the top item at the bottom of the reversed stack 
    
    return
```

```python
def pushAtBottom(s,item_to_be_pushed_boottom):
    
    if s.size()== 0 :
        s.push(item_to_be_pushed_boottom)
        return
    
    current_top = s.pop()
    pushAtBottom(s,item_to_be_pushed_boottom)
    s.push(current_top)
    
    return
```

### Reversing without using any additional space and in O(n) time 

        1. Will have to use the LL as the stack data structure definition 

```python
class StackNode:
     
    def __init__(self, data):
         
        self.data = data
        self.next = None

class Stack:
     
    def __init__(self):
          
        self.top = None
      
    # Push and pop operations
    def push(self, data):
     
        if (self.top == None):
            self.top = StackNode(data)
            return
         
        s = StackNode(data)
        s.next = self.top
        self.top = s
      
    def pop(self):
     
        s = self.top
        self.top = self.top.next
        return s
  
    # Prints contents of stack
    def display(self):
     
        s = self.top
         
        while (s != None):
            print(s.data, end = ' ')
            s = s.next
         
    # Reverses the stack using simple
    # linked list reversal logic.
    def reverse(self):
 
        prev = self.top
        cur = self.top
        cur = cur.next
        succ = None
        prev.next = None
         
        while (cur != None):
            succ = cur.next
            cur.next = prev
            prev = cur
            cur = succ
         
        self.top = prev
```

```python
s = Stack()
s.push(1)
s.push(2)
s.push(3)
s.push(4)

print("Original Stack")
s.display()
print()

# Reverse
s.reverse()

print("Reversed Stack")
s.display()
```

```python
def camStockSpan(arr):
    
    myStack = []
    myStack.append(0)
    span = [1,]
    
    for i in range(1,len(arr)):
        
        while len(myStack) > 0 and arr[i] > arr[myStack[-1]]:
            myStack.pop()
        
        if len(myStack)==0:
            span.append(i+1)
        else:
            span.append(i-myStack[-1])
            
        myStack.append(i)
        
    return span
```

```python
import math
def noOfReversals(exp):
    
    if len(exp)%2 != 0 :
        return -1
    
    st = []
    openc = 0
    closec = 0
    
    for ele in exp:
        
        if ele == '}' and len(st)!= 0 and st[-1] == '{':
            st.pop()
            openc -= 1
        else:
            if ele =='{':
                openc += 1
            if ele =='}':
                closec += 1
            st.append(ele)
            
    if len(st)%2 != 0 :
        return -1
    
    return math.ceil(openc/2)+ math.ceil(closec/2)
```
